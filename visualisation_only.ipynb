## <span style="color:red">Visualisation</span>

```python
import pickle

def save_dict(dico, filename) :
    with open("dict/"+filename+".pickle", 'wb') as handle:
        pickle.dump(dico, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
def load_dict(filename) :
    with open("dict/"+filename+".pickle", 'rb') as handle:
        return(pickle.load(handle))
```

```python
# Comparaison avant et après lemmatisation 
```

```python
article_titles = load_dict("nostopwords")
dict(list(article_titles.items())[0:1]) # affichage des 5 premiers éléments du dictionnaire
```

```python
article_titles_lemmatized = load_dict("lemmatized")
dict(list(article_titles_lemmatized.items())[0:1]) # affichage des 5 premiers éléments du dictionnaire
```

```python
token_list = list(article_titles_lemmatized.values())
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud #pip install wordcloud
```

```python
# On va générer un nuage de mots à l'aide de la librairy wordcloud
# Pour cela il faut inclure tous les mots ensemble, séparés par un espace
all_words=''
for i in range(len(token_list)):
    all_words=all_words + ' '.join([k for k in token_list[i]])
```

```python
wordcloud = WordCloud(width=800, height=500,random_state=21, max_font_size=110).generate(all_words)
plt.figure(figsize=(15, 12))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off');

plt.savefig('D:/OneDrive/Universite/Master1/Semestre1/2.0___Projet_integre/data_cleaning/after_lemmatization.png',transparent=False)
```

```python
# On va maintenant réaliser un histogramme avec le compte des mots les plus fréquents
```

```python
import seaborn as sns
from collections import Counter

token_list=list(article_titles.values())

all_words=''
for i in range(len(token_list)):
    all_words=all_words + ' '.join([k for k in token_list[i]])

# counts = Counter(all_words) : si on fait on aura la liste des lettres les plus courantes...
all_words_list = all_words.split()
counts = Counter(all_words_list)

common_words = [word[0] for word in counts.most_common(25)]
common_counts = [word[1] for word in counts.most_common(25)]
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(20,12))

sns.barplot(x=common_words,y=common_counts)
plt.title('Mots les plus courants')
plt.savefig('D:/OneDrive/Universite/Master1/Semestre1/2.0___Projet_integre/data_cleaning/after_lemmatization_1.png')
```
